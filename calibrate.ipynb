# Calibrate pipette

tips:
- Advised to first calibrate the pipette via the app, this will make a calibration.json on the raspberry pi, which will always be called first before our own offset. Reason for this, is the blowout function: the plunger reset is always in the calibration of the raspberry pi, not this local function. So to reduce this error, calibrate first via the robot
- home if something is off, gantry can be moved slightly and your calibration will be off
- put this script in root

```python
from opentrons_api.load_save_functions import load_settings
from pendantprop.hardware.opentrons.config import Config
import os

settings = load_settings(file_path="config/settings.json")
config = Config(settings=settings)
left_pipette, right_pipette, containers = config.load_all()

pipette = right_pipette
```

```python
config.home()
```

### set your offset

```python
offset = {
    'x': -0.2,
    'y': 1.6,
    'z': 0.0
}
pipette.move_to_tip_calibrate(well="G11", offset=offset)
```

### move to tip to check offset

```python
pipette.move_to_tip_calibrate(well="A1", offset=offset)
```

```python
pipette.move_to_tip_calibrate(well="A12", offset=offset)
```

```python
pipette.move_to_tip_calibrate(well="H1", offset=offset)
```

```python
pipette.move_to_tip_calibrate(well="H12", offset=offset)
```

### pick up tip

watch out: the offset in the settings will be used

```python
well = "H1"
pipette.pick_up_tip(well=well)
```

```python
pipette.return_tip(well=well)
```
